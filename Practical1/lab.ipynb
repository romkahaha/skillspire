### Q3: Generate a listing of all purchases made by customers.

```python
sql_query4 = """
SELECT CUS_FNAME,	CUS_LNAME,	P_DESCRIPT,	P_PRICE
FROM customer
JOIN invoice on customer.CUS_CODE = invoice.CUS_CODE
JOIN line on invoice.INV_NUMBER = line.INV_NUMBER
JOIN product on LINE.P_CODE = product.P_CODE
"""

df = pd.read_sql_query(sql_query4, conn)
df

```

### Q4: Generate a listing of customer purchases, including the subtotals for each of the invoice line numbers.

```python
sql_query5 = """select CUS_FNAME, CUS_LNAME,	V_NAME,	P_DESCRIPT,	invoice.INV_NUMBER,	INV_DATE,	LINE_UNITS, LINE_UNITS * LINE_PRICE as subtotal
FROM customer
JOIN invoice on customer.CUS_CODE = invoice.CUS_CODE
JOIN line on invoice.INV_NUMBER = line.INV_NUMBER
JOIN product on LINE.P_CODE = product.P_CODE
JOIN vendor on product.V_CODE = vendor.V_CODE

"""

df = pd.read_sql_query(sql_query5, conn)
df
```

### Q5: List the balance characteristics of the customers who have made purchases during the current invoice cycle - that is, for the customers who appear in the INVOICE table.


```python
sql_query6 = """SELECT CUS_FNAME,	CUS_LNAME,	INV_NUMBER,	INV_DATE,	CUS_BALANCE,	customer.CUS_CODE
FROM customer
JOIN invoice on customer.CUS_CODE = invoice.CUS_CODE
  WHERE
       customer.CUS_CODE IN (SELECT DISTINCT invoice.CUS_CODE FROM Invoice)
   ORDER BY
       customer.CUS_CODE;
"""

df = pd.read_sql_query(sql_query6, conn)
df

```
